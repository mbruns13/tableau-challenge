```python
import pandas as pd
```

```python
sep2019 = pd.read_csv("Resources/201909-citibike-tripdata.csv")
sep2019["month"] = "September"
# sep2019.head()

oct2019 = pd.read_csv("Resources/201910-citibike-tripdata.csv")
oct2019["month"] = "October"
# oct2019.head()

nov2019 = pd.read_csv("Resources/201911-citibike-tripdata.csv")
nov2019["month"] = "November"
# nov2019.head()

sep2020 = pd.read_csv("Resources/202009-citibike-tripdata.csv")
sep2020["month"] = "September"

oct2020 = pd.read_csv("Resources/202010-citibike-tripdata.csv")
oct2020["month"] = "October"

nov2020 = pd.read_csv("Resources/202011-citibike-tripdata.csv")
nov2020["month"] = "November"
```

```python
fall_dfs = [sep2019, oct2019, nov2019, sep2020, oct2020, nov2020]

fall_tripdata = pd.concat(fall_dfs)

fall_tripdata.head()
```

```python
fall_tripdata.to_csv("Resources/citi_bike_fall_2019_2020.csv")
```
